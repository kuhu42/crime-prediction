```python
def store_case_in_es(case_id, text, predicted_crime):
    doc = {
        "case_id": case_id,
        "text": text,
        "predicted_crime": predicted_crime
    }
    es.index(index=index_name, id=case_id, body=doc)
    print(f"Stored Case ID {case_id} - {predicted_crime}")

# Example: Storing cases after prediction
case_id = 1
for text, crime_category in corrected_results_v3:
    store_case_in_es(case_id, text, crime_category)
    case_id += 1
```

```python
# Simulating Elasticsearch response format
def simulate_elasticsearch_response(cases):
    return {
        "took": 10,  # time it took to process the query
        "timed_out": False,
        "_shards": {"total": 1, "successful": 1, "failed": 0},
        "hits": {
            "total": {"value": len(cases), "relation": "eq"},
            "max_score": 1.0,
            "hits": [{"_index": "cases", "_source": case} for case in cases]
        }
    }

# Simulate Elasticsearch response with the retrieved cases
es_response = simulate_elasticsearch_response(retrieved_cases)

# Print the simulated Elasticsearch response
print("\nSimulated Elasticsearch Response:")
print(es_response)
```

```python
# Simulate relevance scoring like Elasticsearch's _score
def search_cases_with_score(query):
    results = []
    for case in cases:
        score = 0
        # Increase score based on how closely the query matches
        if query.lower() in case["crime_type"].lower():
            score += 1
        if query.lower() in case["verdict"].lower():
            score += 1
        if score > 0:
            results.append({"case": case, "score": score})
    return sorted(results, key=lambda x: x["score"], reverse=True)

# Simulate a query with scoring
retrieved_cases_with_score = search_cases_with_score("Assault")
print("\nResults with simulated Elasticsearch scoring:")
for result in retrieved_cases_with_score:
    print(result["case"], "Score:", result["score"])
```
